```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('bmh')
```

Učitavanje podataka

```python
df = pd.read_csv('./chocolate.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df_num = df.select_dtypes(include = ['float64', 'int64'])
df_num.head()
```

Vizualizacija podataka o tvrtkama

```python
company_counts = df.company.value_counts()
company_counts
```

```python
df.company.unique()
```

```python
top_companies = company_counts.head(15)

plt.figure(figsize=(12, 6))
plt.xticks(rotation=90)
plt.title("Top 15 companies by chocolate production")
fig = sns.barplot(x=top_companies.index, y=top_companies)
fig.set(ylabel='no. of chocolate bars produced')
plt.show()
```

Vizualizacija podataka o lokacijama tvrtaka

```python
locations_counts = df.company_location.value_counts()
locations_counts
```

```python
df.company_location.unique()
```
